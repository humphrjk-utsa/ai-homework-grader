```r
library(dplyr)
library(tidyr)

# --- Example dataset (replace with your own complete_data) ---
complete_data <- data.frame(
  OrderID = c(1, 1, 1, 2, 2, 3, 3, 3, 4, 5, 5),
  ProductID = c(10, 11, 12, 10, 13, 14, 15, 10, 14, 16, 17),
  Product_Name = c(
    "Keyboard", "Mouse", "Monitor",
    "Keyboard", "Webcam",
    "Chair", "Desk", "Keyboard",
    "Chair", "Headphones", "Microphone"
  )
)

# --- Step 1: Keep only orders with more than 1 product ---
multi_item_orders <- complete_data %>%
  group_by(OrderID) %>%
  filter(n_distinct(ProductID) > 1) %>%
  select(OrderID, ProductID, Product_Name)

# --- Step 2: Generate product pairs for each order ---
product_combinations <- multi_item_orders %>%
  group_by(OrderID) %>%
  summarise(
    pairs = list({
      items <- sort(unique(Product_Name))
      if (length(items) > 1) {
        as.data.frame(t(combn(items, 2)), stringsAsFactors = FALSE)
      } else {
        NULL
      }
    }),
    .groups = "drop"
  ) %>%
  unnest(cols = c(pairs)) %>%
  rename(Product1 = V1, Product2 = V2)

# --- Step 3: Count frequency of each combination ---
product_combinations <- product_combinations %>%
  group_by(Product1, Product2) %>%
  summarise(n = n(), .groups = "drop") %>%
  arrange(desc(n))

# --- Step 4: Required output for autograding ---
cat("Product Combination Analysis:\n")
cat("Total product combinations found:", nrow(product_combinations), "\n")
if (nrow(product_combinations) > 0) {
  cat("Most frequent combination appears", max(product_combinations$n, na.rm = TRUE), "times\n")
}

# --- Step 5: Optional — see top combinations ---
head(product_combinations, 10)

```

```r
library(dplyr)

# --- Ensure complete_data exists (for testing only) ---
# Remove this block if your dataset is already loaded
complete_data <- data.frame(
  Supplier_ID = c(1, 1, 2, 2, 3, 3, 3, 4),
  Supplier_Name = c("TechSource", "TechSource", "CompuWorld", "CompuWorld", "OfficePro", "OfficePro", "OfficePro", "ElectroHub"),
  ProductID = c(10, 11, 12, 13, 14, 15, 16, 17),
  Quantity = c(100, 150, 120, 80, 200, 160, 90, 50),
  Unit_Price = c(10, 20, 15, 25, 12, 18, 10, 40)
)

# --- REQUIRED variable name: critical_suppliers ---
# Analyze supplier performance: revenue contribution + product diversity

critical_suppliers <- complete_data %>%
  group_by(Supplier_ID, Supplier_Name) %>%
  summarise(
    Total_Revenue = sum(Quantity * Unit_Price, na.rm = TRUE),
    Product_Count = n_distinct(ProductID),
    .groups = "drop"
  ) %>%
  # Combine both metrics into a "Criticality_Score"
  mutate(
    Revenue_Share = Total_Revenue / sum(Total_Revenue),
    Diversity_Share = Product_Count / sum(Product_Count),
    Criticality_Score = (Revenue_Share * 0.7) + (Diversity_Share * 0.3)
  ) %>%
  arrange(desc(Criticality_Score))

# --- Required output for autograding ---
cat("Critical Suppliers Analysis:\n")
cat("Total suppliers analyzed:", nrow(critical_suppliers), "\n")
print("Top 3 most critical suppliers:")
print(head(critical_suppliers, 3))
```

```r
library(dplyr)

# --- Example dataset (remove this section if complete_data is already loaded) ---
# Assuming complete_data includes CustomerID, City, OrderID, and Total_Amount
complete_data <- data.frame(
  CustomerID = rep(1:10, each = 2),
  City = c("Phoenix", "Phoenix", "Chicago", "Chicago", "Houston", "Houston", "LA", "LA", "NYC", "NYC",
           "Phoenix", "Chicago", "Houston", "LA", "NYC", "Chicago", "Houston", "Phoenix", "LA", "NYC"),
  OrderID = 1:20,
  Total_Amount = c(100, 200, 300, 120, 250, 700, 800, 400, 500, 900,
                   150, 100, 600, 700, 300, 500, 450, 800, 200, 400)
)

# --- REQUIRED variable name: market_expansion ---
# Identify cities with multiple customers and high variation in order values

market_expansion <- complete_data %>%
  group_by(City, CustomerID) %>%
  summarise(Avg_Order_Value = mean(Total_Amount, na.rm = TRUE), .groups = "drop") %>%
  group_by(City) %>%
  summarise(
    Customer_Count = n_distinct(CustomerID),
    Avg_City_Order_Value = mean(Avg_Order_Value, na.rm = TRUE),
    Order_Value_SD = sd(Avg_Order_Value, na.rm = TRUE),
    .groups = "drop"
  ) %>%
  filter(Customer_Count > 1 & !is.na(Order_Value_SD)) %>%
  mutate(
    Opportunity_Index = Order_Value_SD * log(Customer_Count + 1)
  ) %>%
  arrange(desc(Opportunity_Index))

# --- Required output for autograding ---
cat("Market Expansion Analysis:\n")
cat("Cities evaluated for expansion:", nrow(market_expansion), "\n")
print("Top expansion opportunities:")
print(head(market_expansion, 5))
```
