```r
# REQUIRED variable name: customer_metrics
# Your code here: Calculate customer lifetime value metrics
# Required columns: CustomerID, Name, Total_Spent, Order_Count, Avg_Order_Value

suppressPackageStartupMessages(library(dplyr))

# Define std_name helper (so the function exists in this cell)
std_name <- function(df, mapping) {
  for (tgt in names(mapping)) {
    cands <- mapping[[tgt]]
    hit <- cands[cands %in% names(df)]
    if (length(hit) > 0 && !(tgt %in% names(df))) {
      names(df)[match(hit[1], names(df))] <- tgt
    }
  }
  df
}

# Normalize column names (no-op if already standard)
orders <- std_name(orders, list(
  OrderID    = c("OrderID","Order_Id","Order_ID","order_id"),
  CustomerID = c("CustomerID","Customer_Id","Customer_ID","customer_id","CustomerId")
))
customers <- std_name(customers, list(
  CustomerID = c("CustomerID","Customer_Id","Customer_ID","customer_id","CustomerId"),
  Name       = c("Name","CustomerName","Customer_Name")
))
order_items <- std_name(order_items, list(
  OrderID    = c("OrderID","Order_Id","Order_ID","order_id"),
  Quantity   = c("Quantity","Qty","quantity"),
  Unit_Price = c("Unit_Price","UnitPrice","Price","UnitPriceUSD")
))

# Build customer_metrics with required columns
customer_metrics <- customers %>%
  select(CustomerID, Name) %>%
  left_join(orders %>% select(OrderID, CustomerID), by = "CustomerID") %>%
  left_join(order_items %>% select(OrderID, Quantity, Unit_Price), by = "OrderID") %>%
  mutate(line_total = coalesce(Quantity, 0) * coalesce(Unit_Price, 0)) %>%
  group_by(CustomerID, Name) %>%
  summarise(
    Total_Spent = sum(line_total, na.rm = TRUE),
    Order_Count = n_distinct(OrderID),
    .groups = "drop"
  ) %>%
  mutate(Avg_Order_Value = ifelse(Order_Count > 0, Total_Spent / Order_Count, 0)) %>%
   arrange(desc(Total_Spent))

# Required output for autograding:
cat("Customer Analysis Summary:
")
cat("Total customers analyzed:", nrow(customer_metrics), "
")
cat("Top customer total spent: $", round(max(customer_metrics$Total_Spent, na.rm = TRUE), 2), "
")
print(head(customer_metrics[order(-customer_metrics$Total_Spent), ], 3))
```

```r
# REQUIRED variable name: product_metrics  
# Your code here: Analyze product performance
# Required columns: ProductID, Product_Name, Total_Revenue, Total_Quantity, Order_Frequency
suppressPackageStartupMessages(library(dplyr))

# Helper: rename to standard column names if variants are present
std_name <- function(df, mapping) {
  for (tgt in names(mapping)) {
    cands <- mapping[[tgt]]
    hit <- cands[cands %in% names(df)]
    if (length(hit) > 0 && !(tgt %in% names(df))) {
      names(df)[match(hit[1], names(df))] <- tgt
    }
  }
  df
}

# Normalize names (no-op if already standard)
products <- std_name(products, list(
  ProductID    = c("ProductID","Product_Id","Product_ID","product_id"),
  Product_Name = c("Product_Name","ProductName","Name","Product")
))
order_items <- std_name(order_items, list(
  OrderID    = c("OrderID","Order_Id","Order_ID","order_id"),
  ProductID  = c("ProductID","Product_Id","Product_ID","product_id"),
  Quantity   = c("Quantity","Qty","quantity"),
  Unit_Price = c("Unit_Price","UnitPrice","Price","UnitPriceUSD")
))
# Aggregate item lines, then left join to all products to include products with zero sales
item_summaries <- order_items %>%
  mutate(line_total = coalesce(Quantity, 0) * coalesce(Unit_Price, 0)) %>%
  group_by(ProductID) %>%
  summarise(
    Total_Revenue   = sum(line_total, na.rm = TRUE),
    Total_Quantity  = sum(Quantity,   na.rm = TRUE),
    Order_Frequency = n_distinct(OrderID),
    .groups = "drop"
  )

product_metrics <- products %>%
  select(ProductID, Product_Name) %>%
  left_join(item_summaries, by = "ProductID") %>%
  mutate(
    Total_Revenue   = coalesce(Total_Revenue, 0),
    Total_Quantity  = coalesce(Total_Quantity, 0),
    Order_Frequency = coalesce(Order_Frequency, 0L)
  ) %>%
  select(ProductID, Product_Name, Total_Revenue, Total_Quantity, Order_Frequency) %>%
  arrange(desc(Total_Revenue))
# Required output for autograding:
cat("Product Analysis Summary:
")
cat("Total products analyzed:", nrow(product_metrics), "
")
cat("Top product revenue: $", round(max(product_metrics$Total_Revenue, na.rm = TRUE), 2), "
")
print(head(product_metrics[order(-product_metrics$Total_Revenue), ], 3))
```
