```r
# REQUIRED variable name: product_combinations
# Your code here: Find products frequently bought together
# Hint: Group by OrderID, filter orders with multiple items

product_combinations <- order_items %>%
  left_join(products %>% select(ProductID, Product_Name), by = "ProductID") %>%
  group_by(OrderID) %>%
  filter(n() > 1) %>%
  summarise(
    items = list(distinct(tibble(ProductID = ProductID, Product_Name = Product_Name))),
    .groups = "drop"
  ) %>%
  mutate(
    pairs = map(items, ~ {
      it <- .x
      if (nrow(it) < 2) return(tibble())
      cmb <- combn(seq_len(nrow(it)), 2)
      tibble(
        ProductA_ID = it$ProductID[cmb[1,]],
        ProductA_Name = it$Product_Name[cmb[1,]],
        ProductB_ID = it$ProductID[cmb[2,]],
        ProductB_Name = it$Product_Name[cmb[2,]]
      )
    })
  ) %>%
  select(pairs) %>%
  unnest(pairs) %>%
  mutate(
    A_id = pmin(ProductA_ID, ProductB_ID),
    B_id = pmax(ProductA_ID, ProductB_ID),
    A_name = ifelse(ProductA_ID == A_id, ProductA_Name, ProductB_Name),
    B_name = ifelse(ProductB_ID == B_id, ProductB_Name, ProductA_Name)
  ) %>%
  group_by(A_id, A_name, B_id, B_name) %>%
  summarise(Count = n(), .groups = "drop") %>%
  arrange(desc(Count)) %>%
  rename(
    ProductA_ID = A_id,
    ProductA_Name = A_name,
    ProductB_ID = B_id,
    ProductB_Name = B_name
  )

# Required output for autograding (uncomment after completing your analysis):
cat("Product Combination Analysis:\n")
cat("Total product combinations found:", nrow(product_combinations), "\n")
if(nrow(product_combinations) > 0) {
  cat("Most frequent combination appears", max(product_combinations$n, na.rm = TRUE), "times\n")
}
```
